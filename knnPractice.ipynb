```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as  plt
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
model = KNeighborsClassifier(n_neighbors=400)
```

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
model.fit(x,y)
```

```python
new_data = [[30,70]]
prediction = model.predict(new_data)
print(" Result :",  prediction[0])
```

```python
plt.figure(figsize=(8, 5))
for i in range(len(x)):
    age = x.iloc[i, 0]
    estimated_salary = x.iloc[i, 1]
    purchased = y.iloc[i]

    if purchased == 1:
        plt.scatter(age, estimated_salary, color='green', label='Purchased' if i == 0 else "")
    else:
        plt.scatter(age, estimated_salary, color='red', label='Not Purchased' if i == 0 else "" )
new_age = new_data[0][0]
new_salary = new_data[0][1]
predicted_purchased = prediction[0]

new_data_color = 'blue'
new_data_label = f'New Data ({new_age}, {new_salary}) - Predicted: {"Purchased" if predicted_purchased == 1 else "Not Purchased"}'

plt.scatter(new_age, new_salary, color=new_data_color, marker='x', s=100, label=new_data_label)


plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('KNN - Purchase Prediction based on Age and Estimated Salary')
plt.legend()
plt.grid(True)
plt.show()
```
